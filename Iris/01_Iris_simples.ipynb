```python
import pandas as pd

base = pd.read_csv('iris.csv')
base
```

```python
previsores=base.iloc[:, 0:4].values
previsores
```

```python
#coluna 4
classe= base.iloc[:, 4].values
classe
```

Convertendo os valores categÃ³ricos para valores numÃ©ricos da *Classe*


```python
from sklearn.preprocessing import LabelEncoder

labelencoder=LabelEncoder()
classe=labelencoder.fit_transform(classe)
classe
```

iris setosa :   1 0 0

iris virginica: 0 1 0

iris versicolor: 0 0 1

```python
from keras.utils import np_utils

classe_dummy=np_utils.to_categorical(classe)
classe_dummy
```

Dividindo a base de dados em treinamento e teste

```python
from sklearn.model_selection import train_test_split

previsores_treinamento,previsores_teste,classe_treinamento,classe_teste=train_test_split(previsores,classe_dummy,test_size=0.25)
```

ConstruÃ§Ã£o da estrutura da rede neural

units=(4 entradas + 3 saidas)/2 =3.5=4

```python
from keras.models import Sequential
from keras.layers import Dense

classificador=Sequential()
#Primeira camada
classificador.add(Dense(units=4, activation='relu',input_dim=4))
#Segunda camada
classificador.add(Dense(units=4, activation='relu'))
#Camada de saÃ­da
#softmax: para problema de classificaÃ§Ã£o com mais de duas classes
#outro exemplo de metrica: kullback_leiber_divergency
classificador.add(Dense(units=3, activation='softmax'))
classificador.compile(optimizer='adam', loss='categorical_crossentropy',
                      metrics='categorical_accuracy')
```

```python
classificador.fit(previsores_treinamento,classe_treinamento, batch_size=10, epochs=1000)
```

Resultado da base de dados de teste


```python
resultado=classificador.evaluate(previsores_teste,classe_teste)
resultado
```

```python
previsoes=classificador.predict(previsores_teste)
previsoes
```

```python
previsoes=(previsoes>0.5)
previsoes
```
