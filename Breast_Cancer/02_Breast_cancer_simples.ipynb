**units** : quantos neurônios farão parte da camada oculta

**fórmula utilizada para definir a quantidade de neurônios de ponto de partida**: 30 (entrada de dados) + 1 (uma saída por ser um problema binário, onde o Maligno e Benigno são 1 e 0 respectivamente) /2 => *(30+1)/2=15.5 =16*

**input_dim** : Quantos elementos existem na camada de entrada (30 atributos previsores)

Os inicializadores definem a maneira de definir os pesos aleatórios iniciais das camadas Keras.


Mais inicializadores:
https://keras.io/api/layers/initializers/

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
classificador=Sequential()
#Primeira camada oculta
#Definição da primeira camada de entrada (input_dim)
classificador.add(Dense(units =16, activation= 'relu', kernel_initializer='random_uniform', input_dim=30)) 
```

```python
#Segunda camada
classificador.add(Dense(units =16, activation= 'relu', kernel_initializer='random_uniform')) 
```

```python
#Camada de saída
#Retorna um neuronio só (0 ou 1)
#Activation=sigmoid pq ela retorna valor 0 ou 1
classificador.add(Dense(units =1, activation= 'sigmoid')) 
```

**Optimizer**: Qual será a função a ser usada para fazer o ajuste dos pesos.

Mais otimizadores:
https://keras.io/api/optimizers/

**Loss**: Função de perda, como será feito o tratamento do erro

Mais otimizadores:
https://keras.io/api/losses/

**Metrics**: função usada para julgar o desempenho do seu modelo.

Mais métricas:
https://keras.io/api/metrics/

**Batch_size**: Vai calcular o erro para n registros de cada vez e atualizar o peso

**Epochs**: Quantas vezes o ajuste dos pesos vai ser feito

Treinando e testando com a base de treinamento

**decay**= valor que vai decrementar

**clipvalue**= para evitar que o valor fique muito disperso dentro do gradiente

```python
from tensorflow import keras
otimizador= keras.optimizers.Adam(learning_rate=0.001, decay=0.0001, clipvalue=0.5)
#classificador.compile(optimizer='adam', loss='binary_crossentropy', metrics=['binary_accuracy'])
classificador.compile(optimizer=otimizador, loss='binary_crossentropy', metrics=['binary_accuracy'])

classificador.fit(previsores_treinamento,classe_treinamento,batch_size=10, epochs=100)
```
